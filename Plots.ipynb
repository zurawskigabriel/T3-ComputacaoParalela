```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
# ── Dados hardcoded ──────────────────────────────────────────────────────────
nucleos         = [1, 3, 7, 15, 31]
tempoExecucao   = [11, 8, 7, 8, 5]
speedUpIdeal    = [1, 2, 4, 8, 16, 32]

# ── Cálculo do Speed-Up e da Eficiência ──────────────────────────────────────
speedUp    = []
eficiencia = []

tempoBase = tempoExecucao[0]

for i in range(len(nucleos)):
    valorSpeedUp    = tempoBase / tempoExecucao[i]
    valorEficiencia = valorSpeedUp / nucleos[i]
    speedUp.append(round(valorSpeedUp, 2))
    eficiencia.append(round(valorEficiencia, 2))

# ── Figura com eixo y secundário ─────────────────────────────────────────────
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Barras – Eficiência (eixo y secundário)
fig.add_trace(
    go.Bar(
        x           = nucleos,
        y           = eficiencia,
        name        = "Eficiência",
        marker_color= "rgba(99, 180, 230, 0.55)",
        marker_line = dict(color="rgba(99, 180, 230, 0.9)", width=1.5),
        width       = [0.35] * len(nucleos),
    ),
    secondary_y=True,
)

# Linha – Speed-Up Ideal
fig.add_trace(
    go.Scatter(
        x         = nucleos,
        y         = speedUpIdeal,
        mode      = "lines",
        name      = "Speed-Up Ideal",
        line      = dict(color="#E04B4B", width=2, dash="dash"),
        opacity   = 0.85,
    ),
    secondary_y=False,
)

# Linha – Speed-Up Obtido
fig.add_trace(
    go.Scatter(
        x         = nucleos,
        y         = speedUp,
        mode      = "lines+markers+text",
        name      = "Speed-Up Obtido",
        line      = dict(color="#1A73E8", width=2.5),
        marker    = dict(size=9, symbol="circle", color="#1A73E8",
                         line=dict(color="white", width=1.5)),
        text      = speedUp,
        textposition = "top center",
        textfont  = dict(size=11, color="#1A73E8"),
    ),
    secondary_y=False,
)

# ── Layout ───────────────────────────────────────────────────────────────────
fig.update_layout(
    title=dict(
        text      = "<b>Escalonamento Forte (árvore não balanceada)</b>",
        font      = dict(size=22, color="#1a1a2e"),
        x         = 0.5,
        xanchor   = "center",
    ),
    plot_bgcolor  = "#f7f9fc",
    paper_bgcolor = "#ffffff",
    legend        = dict(
        orientation = "h",
        yanchor     = "bottom",
        y           = 1.02,
        xanchor     = "right",
        x           = 1,
        bgcolor     = "rgba(255,255,255,0.85)",
        bordercolor = "#cccccc",
        borderwidth = 1,
    ),
    xaxis=dict(
        title     = dict(text="<b>Número de Núcleos</b>", font=dict(size=14)),
        tickmode  = "array",
        tickvals  = nucleos,
        ticktext  = [str(n) for n in nucleos],
        gridcolor = "#e0e0e0",
        showgrid  = True,
    ),
    margin = dict(t=80, b=60, l=60, r=60),
    hovermode = "x unified",
)

# Eixo y primário – Speed-Up
fig.update_yaxes(
    title_text  = "<b>Speed-Up</b>",
    gridcolor   = "#e0e0e0",
    showgrid    = True,
    secondary_y = False,
)

# Eixo y secundário – Eficiência
fig.update_yaxes(
    title_text  = "<b>Eficiência</b>",
    range       = [0, 1.3],
    tickformat  = ".1f",
    showgrid    = False,
    secondary_y = True,
)

fig.show()

# ── Tabela no terminal ───────────────────────────────────────────────────────
cabecalho = (
    f"{'Núcleos':>8} | {'T. Execução (s)':>16} | "
    f"{'Speed-Up':>9} | {'S. Ideal':>9} | {'Eficiência':>11}"
)
separador = "-" * len(cabecalho)

print("\n  ESCALONAMENTO FORTE (árvore não balanceada) – Multiplicação de Matrizes (MPI)\n")
print(separador)
print(cabecalho)
print(separador)

for i in range(len(nucleos)):
    linha = (
        f"{nucleos[i]:>8} | "
        f"{tempoExecucao[i]:>16} | "
        f"{speedUp[i]:>9.2f} | "
        f"{speedUpIdeal[i]:>9} | "
        f"{eficiencia[i]:>11.2f}"
    )
    print(linha)

print(separador)
```
